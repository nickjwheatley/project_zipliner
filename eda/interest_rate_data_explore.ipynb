```python
import pandas_datareader.data as web
import datetime as dt
import pandas as pd

pd.options.display.max_rows = 100
```

```python
start = dt.datetime(2010, 1, 1)
end = dt.datetime.today().date()

ir_30y_mort = web.DataReader('MORTGAGE30US', 'fred', start, end)
ir_ff = web.DataReader('FEDFUNDS', 'fred', start, end)
pce96 = web.DataReader('PCEC96', 'fred', start, end)
pce = web.DataReader('PCE', 'fred', start, end)

pce96['PCEC96_MOM'] = pce96['PCEC96'] / pce96['PCEC96'].shift(1) -1
pce96['PCEC96_YOY'] = pce96['PCEC96'] / pce96['PCEC96'].shift(12) -1

pce['PCE_MOM'] = pce['PCE'] / pce['PCE'].shift(1) -1
pce['PCE_YOY'] = pce['PCE'] / pce['PCE'].shift(12) -1
```
