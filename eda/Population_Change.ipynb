```python
import requests
import pandas as pd
import csv

def process_population_change_data(file_path):
    population_change_dict = {}
    with open(file_path, 'r', newline='', encoding='utf-8') as csvfile:
        csv_reader = csv.reader(csvfile)
        next(csv_reader, None)
        for row in csv_reader:
            if len(row) == 2:
                population_change_dict[row[0]] = row[1]
    return population_change_dict

def process_year_data(year):
    file_path = f'ACSST5Y{year}.S0701-Data.csv'
    meta_data_file_path = 'ACSST5Y2018.S0701-Column-Metadata.csv'

    df = pd.read_csv(file_path)
    population_change_dict = process_population_change_data(meta_data_file_path)
    print('population_change_dict', population_change_dict)
    df = df.rename(columns=population_change_dict)

    filtered_columns = [
        col for col in df.columns
        if "Moved" in col or "Geography" in col or "Geographic Area Name" in col
    ]

    df = df[filtered_columns]
    df = df.iloc[1:]
    # df['Geographic Area Name'] = df['Geographic Area Name'].str.replace('ZCTA5', '', regex=False).str.strip()

    df['Year'] = year

    return df

years = range(2018, 2023)
dfs = [process_year_data(year) for year in years]

combined_df = pd.concat(dfs, ignore_index=True)

display(combined_df.head())

# combined_df.to_csv('combined_housing_data_2018_2022.csv', index=False)

# print(f"\nTotal number of rows: {len(combined_df)}")
print(f"Years included: {combined_df['Year'].unique()}")
# print(f"\nColumns in the dataset:")
# for col in combined_df.columns:
#     print(col)
```
