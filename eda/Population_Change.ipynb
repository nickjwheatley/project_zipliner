```python
import requests
import pandas as pd
import csv
import numpy as np

def process_population_change_data(file_path):
    population_change_dict = {}
    with open(file_path, 'r', newline='', encoding='utf-8') as csvfile:
        csv_reader = csv.reader(csvfile)
        next(csv_reader, None)
        for row in csv_reader:
            if len(row) == 2:
                population_change_dict[row[0]] = row[1]
    return population_change_dict

def process_year_data(year):
    file_path = f'ACSST5Y{year}.S0701-Data.csv'
    meta_data_file_path = 'ACSST5Y2018.S0701-Column-Metadata.csv'

    df = pd.read_csv(file_path)
    population_change_dict = process_population_change_data(meta_data_file_path)
    print('population_change_dict', population_change_dict)
    df = df.rename(columns=population_change_dict)

    filtered_columns = [
        col for col in df.columns
        if "Moved" in col or "Geography" in col or "Geographic Area Name" in col
    ]

    df = df[filtered_columns]
    df = df.iloc[1:]
    # df['Geographic Area Name'] = df['Geographic Area Name'].str.replace('ZCTA5', '', regex=False).str.strip()

    df['Year'] = year

    return df

years = range(2018, 2023)
dfs = [process_year_data(year) for year in years]

combined_df = pd.concat(dfs, ignore_index=True)

display(combined_df.head())

# combined_df.to_csv('combined_housing_data_2018_2022.csv', index=False)

# print(f"\nTotal number of rows: {len(combined_df)}")
print(f"Years included: {combined_df['Year'].unique()}")
# print(f"\nColumns in the dataset:")
# for col in combined_df.columns:
#     print(col)
```

```python
def clean_numeric(x):
    if isinstance(x, str):
        x = x.replace(',', '')
        if '-' in x:
            x = x.split('-')[0]
        x = ''.join(c for c in x if c.isdigit() or c == '.')
    return x

def process_population_data(df):
    columns_to_keep = {
        'Geographic Area Name': 'Area',
        'Year': 'Year',
        'Estimate!!Moved; within same county!!Population 1 year and over': 'Moved_Within_County',
        'Estimate!!Moved; from different county, same state!!Population 1 year and over': 'Moved_Within_State',
        'Estimate!!Moved; from different  state!!Population 1 year and over': 'Moved_From_Other_State',
        'Estimate!!Moved; from abroad!!Population 1 year and over': 'Moved_From_Abroad',
        'Estimate!!Moved; within same county!!Population 1 year and over!!AGE!!25 to 34 years': 'Age_25_34_Within_County',
        'Estimate!!Moved; within same county!!Population 1 year and over!!AGE!!35 to 44 years': 'Age_35_44_Within_County',
        'Estimate!!Moved; within same county!!Population 1 year and over!!AGE!!45 to 54 years': 'Age_45_54_Within_County',
        'Estimate!!Moved; from different  state!!Population 1 year and over!!AGE!!25 to 34 years': 'Age_25_34_From_Other_State',
        'Estimate!!Moved; from different  state!!Population 1 year and over!!AGE!!35 to 44 years': 'Age_35_44_From_Other_State',
        'Estimate!!Moved; from different  state!!Population 1 year and over!!AGE!!45 to 54 years': 'Age_45_54_From_Other_State',
        'Estimate!!Moved; within same county!!INDIVIDUAL INCOME IN THE PAST 12 MONTHS (IN 2018 INFLATION-ADJUSTED DOLLARS)!!Population 15 years and over!!Median income (dollars)': 'Median_Income',
        'Estimate!!Moved; within same county!!INDIVIDUAL INCOME IN THE PAST 12 MONTHS (IN 2018 INFLATION-ADJUSTED DOLLARS)!!Population 15 years and over!!$75,000 or more': 'Income_75k_Plus',
        'Estimate!!Moved; within same county!!POVERTY STATUS IN THE PAST 12 MONTHS!!Population 1 year and over for whom poverty status is determined!!Below 100 percent of the poverty level': 'Below_Poverty_Level',
        'Estimate!!Moved; within same county!!EDUCATIONAL ATTAINMENT!!Population 25 years and over!!Bachelor\'s degree': 'Bachelors_Degree',
        'Estimate!!Moved; within same county!!EDUCATIONAL ATTAINMENT!!Population 25 years and over!!Graduate or professional degree': 'Graduate_Degree',
        'Estimate!!Moved; within same county!!HOUSING TENURE!!Population 1 year and over in housing units!!Householder lived in owner-occupied housing units': 'Owner_Occupied',
        'Estimate!!Moved; within same county!!HOUSING TENURE!!Population 1 year and over in housing units!!Householder lived in renter-occupied housing units': 'Renter_Occupied',
        'Estimate!!Moved; within same county!!Population 1 year and over!!RACE AND HISPANIC OR LATINO ORIGIN!!One race!!White': 'White',
        'Estimate!!Moved; within same county!!Population 1 year and over!!RACE AND HISPANIC OR LATINO ORIGIN!!One race!!Black or African American': 'Black',
        'Estimate!!Moved; within same county!!Population 1 year and over!!RACE AND HISPANIC OR LATINO ORIGIN!!One race!!Asian': 'Asian',
        'Estimate!!Moved; within same county!!Population 1 year and over!!RACE AND HISPANIC OR LATINO ORIGIN!!Hispanic or Latino origin (of any race)': 'Hispanic_Latino'
    }

    df_selected = df[columns_to_keep.keys()].rename(columns=columns_to_keep)

    numeric_columns = df_selected.columns.drop(['Area', 'Year'])
    for col in numeric_columns:
        df_selected[col] = df_selected[col].apply(clean_numeric).replace('', np.nan).astype(float)

    df_selected['Net_Population_Change'] = df_selected['Moved_Within_County'] + df_selected['Moved_Within_State'] + df_selected['Moved_From_Other_State'] + df_selected['Moved_From_Abroad']

    def safe_divide(a, b):
        return np.where(b != 0, a / b, np.nan)

    df_selected['Pct_Young_Adults'] = safe_divide(df_selected['Age_25_34_Within_County'] + df_selected['Age_25_34_From_Other_State'], df_selected['Net_Population_Change'])
    df_selected['Pct_Middle_Aged_Adults'] = safe_divide(df_selected['Age_35_44_Within_County'] + df_selected['Age_35_44_From_Other_State'] + df_selected['Age_45_54_Within_County'] + df_selected['Age_45_54_From_Other_State'], df_selected['Net_Population_Change'])
    df_selected['Pct_Higher_Education'] = safe_divide(df_selected['Bachelors_Degree'] + df_selected['Graduate_Degree'], df_selected['Net_Population_Change'])
    df_selected['Owner_Renter_Ratio'] = safe_divide(df_selected['Owner_Occupied'], df_selected['Renter_Occupied'])

    return df_selected

processed_df = process_population_data(df)

display(f"\nProcessed dataset shape: {processed_df.shape}")
display("\nColumns in the processed dataset:")
display(processed_df.columns.tolist())
display("\nSample of the processed data:")
display(processed_df.head())

numeric_columns = processed_df.columns.drop(['Area', 'Year'])
non_numeric = processed_df[numeric_columns].applymap(lambda x: not np.issubdtype(type(x), np.number))
if non_numeric.any().any():
    print("\nWarning: Some non-numeric data remains:")
    print(non_numeric.sum())
else:
    print("\nAll numeric columns have been successfully converted to float.")
```
